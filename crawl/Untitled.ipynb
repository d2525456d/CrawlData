```python
#import thư viện
import pandas as pd
```

```python
#Có nhiều league và nhiều mùa giải em chỉ lấy EPL mùa giải 2020/2021 để phân tích
#Đưa dữ liệu vào 
File_crawl = pd.read_csv('File_craw4446.csv')
df = File_crawl.copy()
```

```python
#drop 1 cột index dư
df.drop(df.columns[[0]], axis = 1, inplace=True)
df
```

```python
# xem thông tin data
df.info()
```

```python
#Thêm cột Home_point và Away_point vào df để tính tổng điểm của từng đội
Home_point = []
Away_point = []
for i in range(len(df['Home_score'])):
    if df['Home_score'][i]>df['Away_score'][i]:
        Home_point.append(3)
        Away_point.append(0)
    elif df['Home_score'][i]==df['Away_score'][i]:
        Home_point.append(1)
        Away_point.append(1)
    elif df['Home_score'][i]<df['Away_score'][i]:
        Home_point.append(0)
        Away_point.append(3)
df.insert(6,"Home_point",Home_point)
df.insert(7,"Away_point",Away_point)
df
```

```python
#Lấy 2 cột Home_team và Home_point ra
Home = df[['Home_team','Home_point']]
Home.columns = range(Home.shape[1])
Home
```

```python
#Tương tự như trên
Away = df[['Away_team','Away_point']]
Away.columns = range(Away.shape[1])
Away
```

```python
#gom 2 cột Home và Away lại
HA = pd.concat([Home,Away])
HA.rename(columns = {0:'Team',1:'Point'}, inplace=True)
HA
```

```python
#tổng lại điểm của từng đội
BXH = HA.groupby(["Team"]).sum()
BXH = BXH.reset_index()
BXH
```

```python
#thêm 2 cột GD_home và GD_away để tính hệ số
GD_home = []
GD_away = []
for i in range(len(df['Home_score'])):
    GD_home.append(df['Home_score'][i]-df['Away_score'][i])
    GD_away.append(df['Away_score'][i]-df['Home_score'][i])
df.insert(8,"GD_home",GD_home)
df.insert(9,"GD_away",GD_away)
df
```

```python
#Tách 2 cột Home_team và GD_home
Home_GD = df[['Home_team','GD_home']]
Home_GD.columns = range(Home_GD.shape[1])
Home_GD
```

```python
#tương tự như trên
Away_GD = df[['Away_team','GD_away']]
Away_GD.columns = range(Away_GD.shape[1])
Away_GD
```

```python
#Gom Home_GD và Away_GD lại
HA_GD = pd.concat([Home_GD,Away_GD])
HA_GD.rename(columns = {0:'Team',1:'GD'}, inplace=True)
HA_GD
```

```python
#tính tổng hiệu số của từng đội
T_GD = HA_GD.groupby(["Team"]).sum()
T_GD = T_GD.reset_index()

T_GD
```

```python
#thêm cột GD vào BXH
BXH['GD'] = T_GD['GD']
BXH
```
